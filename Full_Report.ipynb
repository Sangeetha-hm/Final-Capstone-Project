<h2>Introduction</h2>
<hr>
<b>Identifying the busiest neighborhoods for a ride-hailing service venture</b>

A new ride-hailing service venture is being established in the city of New York. The company is looking for predictive demand analysis in Real-Time for fleet management and optimization. The goal of this project is to analyze footfalls in the popular venues across the neighborhoods of the city to project potential demand for taxis. Based on these projections, the business plans to direct their fleet accordingly for just-in-time servicing of demand.

<h2>Methodology</h2>
<hr>

The methodology followed in this project can be broken down into 4 stages.
1. Data Collection
2. Data Cleaning
3. Machine Learning
4. Data Visualization

<b>Stage 1: Collecting the data</b>
This stage involves gathering the data from the sources identified in the Data description section. The datasets used in this project are retrieved from two main sources. One is the [Newyork government website](https://www.health.ny.gov/statistics/cancer/registry/appendix/neighborhoods.htm) and the other is from the [Foursquare API](https://developer.foursquare.com/docs/api/venues/trending).

<b>Stage 2: Cleaning the data</b>
This is a rather time consuming stage. The Newyork neighborhoods data collected in the previous stage is in the  json format. It is then processed to get the data in a tabular format and written into a csv file. The notebook for the Newyork data is available [here](https://github.com/Sangeetha-hm/Final-Capstone-Project/blob/master/Notebooks/Newyork_Postcodes.ipynb). <b>Stage 3: Machine Learning</b>
K-means clustering algorithm is used on the trending venues dataframe for identifying the probable clusters for the busiest neighborhoods at any given time.

<b>Stage 4: Visualizing the data</b>
Folium is used for data visualization. Two maps are generated, one for visualizing the trending venues across the New York city and the other for clusters generated by the k-means algorithm superimposed on the New York city map.

<h2>Discussion and Conclusion</h2>
<hr>
1. The traditional approach of basing demand on historically popular venues would lead to missing, for example, an event like the 'New York Woman's March' in Upper West Side Manhattan that comes up at number 10 in the trending venues.
<br><br>
2. Also, the traditional approach does not make allowances, for example, a popular movie release driving more footfall to the movie halls on a given day.
<br><br>
3. Even the approach of just depending on a list of trending venues (before clustering and visualization) shows venues spread across the city of New York, making it difficult for a fleet management company to draw any insights from.
<br><br>
4. Chelsea & Clinton neighborhood does not feature much in the list of top 10 trending venues, although it features with the densest hotspot after the ML algorithm is applied, which makes it the best target neighborhood.
<br><br>
5. The company can, with the help of the visualization, even drill down to the areas in a neighborhood that feature the denser groupings of hotspots, making it easier to target potential demand.
<br><br>
6. The category data has also been gleaned, but not made use of in this project; but this data can be used to further quantify demand. For example, demand would be more from footfall out of pubs, nightclubs, etc than from a movie hall. This can be considered in scope for future enhancement of this project.
